```python
import tkinter as tk
from tkinter import Canvas, Button, PhotoImage

class Emg(tk.Frame):
    def __init__(self, master):
        super().__init__(master)

        self.canvas = Canvas(
            self,
            bg="#AFAFAF",
            height=800,
            width=1192,
            bd=0,
            highlightthickness=0,
            relief="ridge",
        )

        self.canvas.place(x=0, y=0)
        self.image_image_1 = PhotoImage(file="assets/emg/image_1.png")
        self.image_1 = self.canvas.create_image(596.0, 342.0, image=self.image_image_1)

        self.button_image_1 = PhotoImage(file="assets/emg/button_1.png")
        self.button_fullscreen = Button(
            self,
            image=self.button_image_1,
            borderwidth=0,
            highlightthickness=0,
            command=self.toggle_fullscreen,
            relief="flat",
        )
        self.button_fullscreen.place(x=323.0, y=619.0, width=546.0, height=69.0)

        self.button_exit_fullscreen = Button(
            self,
            text="Exit Fullscreen",
            command=self.exit_fullscreen,
        )
        self.button_exit_fullscreen.place(x=20, y=20)

        self.fullscreen = False  # Store fullscreen state

    def toggle_fullscreen(self):
        # Toggle fullscreen mode
        self.master.attributes('-fullscreen', not self.fullscreen)
        self.fullscreen = not self.fullscreen

    def exit_fullscreen(self):
        # Exit fullscreen mode
        self.master.attributes('-fullscreen', False)
        self.fullscreen = False

if __name__ == "__main__":
    root = tk.Tk()

    # Set window to initial windowed mode
    root.attributes('-fullscreen', False)

    root.geometry("1280x800")
    
    # Create a container frame
    container_frame = tk.Frame(root)
    container_frame.config(width=1200,height=800)
    container_frame.pack()
    
    # Instantiate Emg class and place it inside the container frame
    frame = Emg(container_frame)
    frame.config(width=1200,height=800)
    frame.pack()
    
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import Tk, Canvas, Entry, Text, Button, PhotoImage
import pandas as pd


class Env(tk.Frame):
    def __init__(self, master):
        super().__init__(master)

        self.canvas = Canvas(
            self,
            bg="#AFAFAF",
            height=800,
            width=1200,
            bd=0,
            highlightthickness=0,
            relief="ridge",
        )

        self.canvas.place(x=0, y=0)
        # self.image_image_1 = PhotoImage(file="assets/env/image_1.png")
        # self.image_1 = self.canvas.create_image(600.0, 678.0, image=self.image_image_1)

        self.tempText = self.canvas.create_text(
            428.0,
            110.0,
            anchor="nw",
            text="Temperature: 21°C ",
            fill="#000000",
            font=("Inter Bold", 36 * -1),
        )

        self.humdText = self.canvas.create_text(
            428.0,
            162.0,
            anchor="nw",
            text="Humidity: 50 % ",
            fill="#000000",
            font=("Inter Bold", 36 * -1),
        )

        self.pressText = self.canvas.create_text(
            428.0,
            214.0,
            anchor="nw",
            text="Pressure: 1000 hPa ",
            fill="#000000",
            font=("Inter Bold", 36 * -1),
        )

        # self.image_image_2 = PhotoImage(file="assets/env/image_2.png")
        # self.image_2 = self.canvas.create_image(600.0, 466.0, image=self.image_image_2)

        self.updatedata()

        self.forecast = tk.Frame(self)
        self.forecast.place(x=600.0, y=678.0)

        df = pd.read_csv("")

        self.monLabel = tk.Label(self.forecast, text = "mon\n75\n0")
        self.monLabel.pack()


    def updatedata(self):
        # if self.loop:
        self.df = pd.read_csv("frames/env/Sensor_data.csv")
        self.temp = self.df.iloc[0, 1]
        self.hum = self.df.iloc[0, 2]
        self.pres = self.df.iloc[0, 3]
        # print(self.temp)
        self.canvas.itemconfig(self.tempText, text=f"Temperature: {self.temp:.2f}°C")
        self.canvas.itemconfig(self.humdText, text=f"Humidity: {self.hum:.2f}%")
        self.canvas.itemconfig(self.pressText, text=f"Pressure: {self.pres:.2f} hPa")
        self.after(2000, self.updatedata)

if __name__ == "__main__":
    root = tk.Tk()
    root.geometry("1280x800")

    # Instantiate Plant class
    frame = Env(root)
    frame.config(width=1200, height=800)

    # Pack the Plant frame
    frame.grid(row=0, column=0)
    root.mainloop()
```

```python
import tkintermapview
import os


# This scripts creates a database with offline tiles.

# specify the region to load (New York City)
top_left_position = (33.94202642664497, -84.52396022497382)
bottom_right_position = (33.934157867287084, -84.5103560618182)
zoom_min = 0
zoom_max = 15

# specify path and name of the database
# script_directory = os.path.dirname(os.path.abspath(__file__))
# database_path = os.path.join(script_directory, "offline_tiles_nyc.db")

# create OfflineLoader instance
loader = tkintermapview.OfflineLoader(path="offline_tiles.db", tile_server="https://mt0.google.com/vt/lyrs=m&hl=en&x={x}&y={y}&z={z}&s=Ga")

# save the tiles to the database, an existing database will extended
loader.save_offline_tiles(top_left_position, bottom_right_position, zoom_min, zoom_max)

# You can call save_offline_tiles() multiple times and load multiple regions into the database.
# You can also pass a tile_server argument to the OfflineLoader and specify the server to use.
# This server needs to be then also set for the TkinterMapView when the database is used.
# You can load tiles of multiple servers in the database. Which one then will be used depends on
# which server is specified for the TkinterMapView.

# print all regions that were loaded in the database
loader.print_loaded_sections()
```
