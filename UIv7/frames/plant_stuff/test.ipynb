```python
import tkinter as tk
from tkinter import Tk, Canvas, Entry, Text, Button, PhotoImage
# import cv2
# import os
# import torch
from PIL import Image, ImageTk


class Plant(tk.Frame):
    def __init__(self, master):
        super().__init__(master)

        self.CameraUpdate = True  # Corrected variable name

        self.canvas = Canvas(
            self,
            bg="#AFAFAF",
            height=800,
            width=1200,
            bd=0,
            highlightthickness=0,
            relief="ridge",
        )

        self.canvas.place(x=0, y=0)

        # midline

        self.midline = Canvas(
            self,
            bg="black",
            height=800,
            width=5,
            bd=0,
            highlightthickness=0,
            relief="ridge",
        )

        self.midline.place(x=600, y=0)


        # capture button

        self.capture_button = tk.Button(
            self,
            text="Capture",
            #command=self.capture_image,
            anchor="center",
        )
        self.capture_button.place(x=self.findCenterx(self.capture_button), y=600)

        # camera_label / image thing

        # self.camera_label = tk.Label(self, text="test", width = 640, height = 480)
        # # self.camera_label.place(x=((1200 - 640) // 2), y=0, )
        # self.camera_label.place(x=0, y=0)

        self.camera_label = tk.Canvas(
            self,
            width=640,
            height=480,
            bd=0,
            highlightthickness=0,
            relief="ridge",
        )
        self.camera_label.place(x=0, y=0)

        image = Image.open("frames/plant_stuff/captured_image.jpg")
        photo = ImageTk.PhotoImage(image)
        self.camera_label.create_image(0, 0, anchor="nw", image=photo)


        # result_label

        self.result_label = tk.Label(
            self, text="test", font=("Helvetica", 12), foreground="black"
        )  # Set text color to black
        self.result_label.place(x=self.findCenterx(self.result_label), y=500)


    def findCenterx(self, object):
        self.object = object
        self.width = 600
        self.objectWidth = object.winfo_reqwidth()

        self.x = self.width - self.objectWidth // 2

        return self.x


if __name__ == "__main__":
    root = tk.Tk()
    root.geometry("1280x800")

    # Instantiate Plant class
    frame = Plant(root)
    frame.config(width=1200, height=800)

    # Pack the Plant frame
    frame.grid(row=0, column=0)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import Tk, Canvas, Entry, Text, Button, PhotoImage
import pandas as pd


class Env(tk.Frame):
    def __init__(self, master):
        super().__init__(master)

        self.canvas = Canvas(
            self,
            bg="#AFAFAF",
            height=800,
            width=1200,
            bd=0,
            highlightthickness=0,
            relief="ridge",
        )

        self.canvas.place(x=0, y=0)
        self.image_image_1 = PhotoImage(file="assets/env/image_1.png")
        self.image_1 = self.canvas.create_image(600.0, 678.0, image=self.image_image_1)

        self.tempText = self.canvas.create_text(
            428.0,
            110.0,
            anchor="nw",
            text="Temperature: 21°C ",
            fill="#000000",
            font=("Inter Bold", 36 * -1),
        )

        self.humdText = self.canvas.create_text(
            428.0,
            162.0,
            anchor="nw",
            text="Humidity: 50 % ",
            fill="#000000",
            font=("Inter Bold", 36 * -1),
        )

        self.pressText = self.canvas.create_text(
            428.0,
            214.0,
            anchor="nw",
            text="Pressure: 1000 hPa ",
            fill="#000000",
            font=("Inter Bold", 36 * -1),
        )

        self.image_image_2 = PhotoImage(file="assets/env/image_2.png")
        self.image_2 = self.canvas.create_image(600.0, 466.0, image=self.image_image_2)

        self.updatedata()

    def updatedata(self):
        # if self.loop:
        self.df = pd.read_csv("frames/env/Sensor_data.csv")
        self.temp = self.df.iloc[0, 1]
        self.hum = self.df.iloc[0, 2]
        self.pres = self.df.iloc[0, 3]
        # print(self.temp)
        self.canvas.itemconfig(self.tempText, text=f"Temperature: {self.temp:.2f}°C")
        self.canvas.itemconfig(self.humdText, text=f"Humidity: {self.hum:.2f}%")
        self.canvas.itemconfig(self.pressText, text=f"Pressure: {self.pres:.2f} hPa")
        self.after(2000, self.updatedata)

if __name__ == "__main__":
    root = tk.Tk()
    root.geometry("1280x800")

    # Instantiate Plant class
    frame = Env(root)
    frame.config(width=1200, height=800)

    # Pack the Plant frame
    frame.grid(row=0, column=0)
    root.mainloop()
```
